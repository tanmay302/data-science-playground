```python
# Columns having outlier
def outlier_summary(df):
    result = {}
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        
        if len(outliers) > 0:
            result[col] = len(outliers)
    
    return result
```

```python
def outlier_summary(df):
    result = {}
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    col_with_outliers=[]
    
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        if df[(df[col] < lower) | (df[col] > upper)].shape[0]>0:
            col_with_outliers.append(col)
        
    return col_with_outliers
```

```python
print("Columns having outliers:",outlier_summary(df))
cols=outlier_summary(df)
```

```python
# From this we understand all values have outlier values in it and we have to remove outlier for good prediction
# Removing outlier
def remove_outlier(df,col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))
    return df
```

```python
df=remove_outlier(df,cols)
df.head()
```

```python
# Calling summary_df again to verify outlier is present or not
if outlier_summary(df):
    print("Outlier Still Present")

else:
    print("Outlier Removed")
```

#### 5.	Conclusion

#### Summary of Key Findings from Exploratory Data Analysis

After performing exploratory data analysis, several useful patterns were observed in the dataset.

First, some variables such as DS and DP contained constant values (no variation across observations). Since they do not change, they do not provide meaningful information for analysis and can be removed.

The correlation analysis showed that a few variability-related features are strongly connected with each other. For example, MSTV and Width showed a strong positive relationship, meaning that when one increases, the other tends to increase as well. Similarly, MSTV and ALTV showed a strong negative relationship, indicating that as one rises, the other tends to decrease.

This suggests that many of the variability features are measuring related aspects of fetal heart rate behavior.

On the other hand, features like LB and FM showed weak relationships with most other variables. These may provide independent information rather than overlapping with other features.

No time-based analysis was performed because the dataset does not contain any temporal variable such as date or timestamp.

#### Implications for Further Analysis

The strong relationships between certain variables indicate possible redundancy. If linear models are used, this could lead to multicollinearity. In such cases, feature selection or regularization techniques may be helpful.

Removing constant features will improve model efficiency and reduce unnecessary complexity.

Since variability-related features appear important and strongly connected, they may play a key role in predictive modeling and should be carefully analyzed when building models.

Overall, these findings help guide feature selection, model choice, and further analysis steps.
