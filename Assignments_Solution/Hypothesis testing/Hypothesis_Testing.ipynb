# Hypothesis Testing

## 1.State the Hypothesis

The theoretical weekly operating cost model predicts:

μ = ₹4000

Since restaurant owners claim that operating costs are higher than predicted,
we conduct a right-tailed test.

### Null Hypothesis (H₀):
The mean weekly operating cost follows the theoretical model.
μ = 4000

### Alternative Hypothesis (H₁):
The mean weekly operating cost is higher than the theoretical model.
μ > 4000

## 2. Calculate the Test Statistic

```python
# Importing required Packages
import math
```

```python
# Given things
x_bar = 3050
mu = 4000
sigma = 125
n = 25

# Calculating z
z = (x_bar - mu) / (sigma / math.sqrt(n))
z
```

## 3. Determine the Critical Value

```python
from scipy.stats import norm

alpha = 0.05
z_critical = norm.ppf(1 - alpha)

z_critical
```

## 4. Make a Decision:


```python
z_calculated = -38
z_critical = 1.645

if z_calculated > z_critical:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

## 5. Conclusion

At the 5% significance level (α = 0.05), we failed to reject the null hypothesis.

Therefore, there is not enough statistical evidence to support the restaurant owners' claim that the weekly operating costs are higher than predicted by the theoretical model.

In fact, the sample mean weekly cost (₹3050) is significantly lower than the theoretical mean cost (₹4000).

Hence, the data does not support the claim that operating costs have increased.
