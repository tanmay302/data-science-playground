```python
# Exercise 1: Prime Numbers

n=int(input("Enter Number to check for prime"))  #Accepting Input
isPrime=True

if n<0:               #If n is negative
    isPrime=False

for i in range(2,n//2+1):  #Logic for Prime Number check
    if(n%i==0):
        isPrime=False
        break;

if isPrime:     
    print(n," is Prime Number")   #If isPrime is still True
else:
    print(n," is not Prime Number")#If isPrime is False
```

```python
# Exercise 2: Product of Random Numbers

# For random number generation we have to import Random Library
import random

# Accpeting Product input from user.
prod=int(input("Enter the product of two randomly generated numbers from 1 to 10:"))

# Generating two random numbers from 1 to 10
random_n1=random.randint(1,10)
random_n2=random.randint(1,10)

# Calculating Product of Randomly generated numbers
random_product=random_n1*random_n2

# Checking user's input vs Calculated Product
if prod==random_product:
    print("Your answer is Correct")
    print("The generated numbers were {0} and {1}".format(random_n1,random_n2))

else:
    print("Your answer is Wrong")
    print("The generated numbers were {0} and {1}".format(random_n1,random_n2))
```

```python
# Exercise 3: Squares of Even/Odd Numbers

# Accepting Even-Odd Choice from User
choice=input("Enter your choice to print squares: Even/Odd =>").lower()

# If User Entered Even
if choice=='even':
    for square in range(100,201):
        if(square%2==0):
            print("Square of {0}: {1}".format(square,square*square))

#If User Entered Odd
elif choice=='odd':
    for square in range(100,201):
        if(square%2!=0):
            print("Square of {0}: {1}".format(square,square*square))

# If User Entered other than even or odd
else:
    print("Invalid Input")
```

```python
# Exercise 4: Word counter

# Accepting text from user
text=input("Enter text:")

# Remove Punctuation and split into words
words=text.replace(".","").replace(",","").split()

# Dictionary to store word counts
word_count={}

for word in words:
    if word in word_count:
        word_count[word]+=1

    else:
        word_count[word]=1

# Printing Result
for word,count in word_count.items():
    print(f"'{word}':'{count}'")
```

```python
# Exercise 5: Check for Palindrome

input_text=input("Enter text for checking Palindrom")

# Remove spaces and convert to lowercase for uniformity
cleaned_text=input_text.replace(" ","").lower()

# Check if the string is equal to its reverse
if cleaned_text == cleaned_text[::-1]:
    print(input_text," is Palindrome")
    
else:
    print(input_text," is Not Palindrome")
```
