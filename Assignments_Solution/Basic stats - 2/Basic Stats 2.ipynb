Estimation And Confidence Intervals

```python
#Importing Packages
import numpy as np
import scipy.stats as stats 
```

```python
# Data
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
        1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
print(data)
```

```python
#Converting data to numpy array
data=np.array(data)
data
```

```python
#Sample Statistics
n=len(data)
mean=np.mean(data)
sample_std=np.std(data,ddof=1)

print("Sample Size:",n)
print("Sample Mean:",round(mean,3))
print("Sample Standard Deviation:",round(sample_std,3))
```

```python
# a. Build 99% Confidence Interval Using Sample Standard Deviation
alpha=0.01
df=n-1
t_critical=stats.t.ppf(1-alpha/2,df)
margin_error_t=t_critical*(sample_std/np.sqrt(n))
ci_t=(mean - margin_error_t, mean + margin_error_t)

print("\n99% CI using sample std (t-distribution):")
print("(", round(ci_t[0], 3), ",", round(ci_t[1], 3), ")")
```

```python
# b. Build 99% Confidence Interval Using Known Population Standard Deviation
pop_std = 0.2
z_critical = stats.norm.ppf(1 - alpha/2)

margin_error_z = z_critical * (pop_std / np.sqrt(n))
ci_z = (mean - margin_error_z, mean + margin_error_z)

print("\n99% CI using population std (z-distribution):")
print("(", round(ci_z[0], 3), ",", round(ci_z[1], 3), ")")
```
