### Scaling Techniques

### Standard Scaling
- Converts data to mean = 0 and standard deviation = 1  
- Use when data is normally distributed  
- Works well with:
  - Linear Regression  
  - Logistic Regression  
  - SVM  
  - PCA  
- Reason:
  - Centers data
  - Reduces effect of large values

---

### Min-Max Scaling
- Scales data between 0 and 1  
- Use when data is not normally distributed  
- Works well with:
  - KNN  
  - Neural Networks  
- Reason:
  - Keeps data in fixed range
  - Preserves relationships between values

---

### Conclusion
- Use Standard Scaling for statistical models  
- Use Min-Max Scaling for distance-based models  

### 2. Encoding Techniques

```python
# Categorical Columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_cols)
```

```python
# Divide Based on Number of Categories
one_hot_cols = [col for col in categorical_cols if df[col].nunique() < 5]
label_cols = [col for col in categorical_cols if df[col].nunique() >= 5]

print("One-Hot Columns:", one_hot_cols)
print("Label Encoding Columns:", label_cols)
```

```python
# Apply Encoding
# One-Hot Encoding
df = pd.get_dummies(df, columns=one_hot_cols)

# Label Encoding
encoders = {}

for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
# Shape after encoding
print("Shape after encoding:", df.shape)

# Preview dataset
print(df.head())

# Check data types
print(df.dtypes)

# Example: show mapping of one label encoded column
col = label_cols[0]
print(f"\nMapping for {col}:")
print(dict(zip(encoders[col].classes_, encoders[col].transform(encoders[col].classes_))))
```

### Pros and Cons of Encoding Techniques

### One-Hot Encoding

**Pros:**
- No order between categories  
- Works well with most machine learning models  

**Cons:**
- Increases number of columns  
- Uses more memory and computation  

---

### Label Encoding

**Pros:**
- Simple and fast  
- Uses less memory  

**Cons:**
- Introduces order between categories  
- Can reduce model accuracy in some cases  

### 3. Feature Engineering

```python
# Create 2 New features
# Feature 1 : Total Capital Gain
df["total_capital"] = df["capital_gain"] - df["capital_loss"]
```

```python
# Feature 2: work_hours_level
def work_hours_category(x):
    if x < 30:
        return "Low"
    elif x <= 50:
        return "Medium"
    else:
        return "High"

df["work_hours_level"] = df["hours_per_week"].apply(work_hours_category)
```

```python
# Transforming Skewed Feature
df["capital_gain_log"] = np.log1p(df["capital_gain"])
```

```python
print(df[["capital_gain", "capital_gain_log"]].head())

print("\nNew Features Added:")
print(df[["total_capital", "work_hours_level"]].head())
```

### Justification of Feature Engineering

### New Features

**1. total_capital**
- Combines capital_gain and capital_loss  
- Represents overall financial status more effectively  
- Helps the model understand net income-related behavior  

**2. work_hours_level**
- Converts numerical hours_per_week into categories (Low, Medium, High)  
- Makes patterns easier for the model to learn  
- Useful for models that perform better with categorical data  

---

### Transformation

**capital_gain (Log Transformation)**
- The feature is highly skewed (many small values and few very large values)  
- Log transformation reduces skewness  
- Helps in improving model performance and stability  
- log1p is used to handle zero values safely  

---

### Conclusion
Feature engineering improves data representation by creating meaningful features and reducing skewness, which helps the model learn better patterns.

### 4. Feature Selection

```python
# Identifying and removing outliers using Isolation Forest Algorithm
# Select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
outliers = iso.fit_predict(df[num_cols])

# Keep only normal data (1 = normal, -1 = outlier)
df = df[outliers == 1]
print("Shape before removing outliers:(32561, 15)")
print("Shape after removing outliers:", df.shape)
```

```python
# Applying PPS (Predictive Power Score) 
pps_matrix = pps.matrix(df)
print(pps_matrix.head())
```

```python
# Correlation Matrix
corr_matrix = df.select_dtypes(['int','float64']).corr()
print(corr_matrix)
```

### Comparison: PPS vs Correlation

### Correlation

- Measures linear relationship between features  
- Works only with numerical data  
- Values range from -1 to 1  
- Fast and easy to compute  

**Limitations:**
- Cannot detect non-linear relationships  
- Ignores categorical variables  

---

### PPS (Predictive Power Score)

- Measures how well one feature predicts another  
- Works with both numerical and categorical data  
- Values range from 0 to 1  
- Can detect non-linear relationships  

**Limitations:**
- Slower to compute  
- Less intuitive than correlation  

---

## Conclusion

- Correlation is useful for identifying linear relationships  
- PPS is more powerful as it detects both linear and non-linear relationships
